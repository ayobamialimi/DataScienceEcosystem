```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# I want to handle the NAN, get the median of the bedrooms
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
# fill the na value
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
df
```

```python
# create linear regression

# call the linear regression model
reg = linear_model.LinearRegression()

# train the model with your dataset
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
`reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[3000,3,5]])
```

# Multivariate regression model on hiring data

```python
df2 = pd.read_csv("hiring.csv")
df2
```

```python
#rename some columns
df2.rename(columns={'test_score(out of 10)': 'test_score', 'interview_score(out of 10)': 'interview_score', 'salary($)':'salary' }, inplace=True)
```

```python
df2
```

```python
# I want to handle the NAN, get the median of the bedrooms
import math
median_testscore = math.floor(df2.test_score.median())
median_testscore
```

```python
# fill the na value

df2.test_score = df2.test_score.fillna(median_testscore)
```

```python
df2.loc[0,'experience']= 'four'
df2.loc[1,'experience']= 'six'
df2
```

```python
# Dictionary mapping string values to numeric values
mapping = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'six': 6, 'seven': 7, 'eight': 8, 'ten': 10, 'eleven': 11}

#replace the string with the numeric values
df2.replace(mapping, inplace=True)
df2
```

```python
# create linear regression

# call the linear regression model
reg = linear_model.LinearRegression()

# train the model with your dataset
reg.fit(df2[['experience', 'test_score', 'interview_score']], df2.salary)
```

```python
reg.predict([[6,8,6]])
```
