```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
# print the first 5 images
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
#plt.show()
```

```python
# Print the actual numbers from 0-5
digits.target[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

Test a sample prediction

```python
# lets try and show a random number
plt.matshow(digits.images[76])
```

```python
digits.target[76]
```

```python
model.predict(digits.data[[76]])
```

```python
for i in range(10,100,10):
  print(model.predict([digits.data[i]]))
```

```python
model.predict(digits.data[0:5])
```

```python
# Confusion matrix
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# better visualization with seaborn
import seaborn as sn
plt.figure(figsize = (10,7))
bn=sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
iris.data[0:5]
```

```python
iris.target[100:150]
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2)
len(X_train)
```

```python
len(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# confusion matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# better visualization with seaborn
import seaborn as sn
plt.figure(figsize = (10,7))
bn=sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
